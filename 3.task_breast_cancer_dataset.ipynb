```python
from sklearn.model_selection import train_test_split
X= data_df.drop("target", axis = 1)
y = data_df["target"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
y_train
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns
svc_linear = SVC(kernel='linear')
svc_linear.fit(X_train, y_train)

```

```python
y_pred_linear = svc_linear.predict(X_test)

accuracy_linear = accuracy_score(y_test, y_pred_linear)
print("Accuracy:", accuracy_linear)
```

```python
cm = confusion_matrix(y_test, y_pred_linear)

sns.heatmap(cm, annot=True, fmt="d", cmap="Blues").set_title("Confusion Matrix")
```

```python
svc_poly = SVC(kernel='poly')
svc_poly.fit(X_train, y_train)
```

```python
y_pred_poly = svc_poly.predict(X_test)

accuracy_poly = accuracy_score(y_test, y_pred_poly)
print("Accuracy:", accuracy_poly)
```

```python
#Radial SVM or Gausian Kernel
svc_gaussian = SVC(kernel='rbf', gamma=0.1)
svc_gaussian.fit(X_train, y_train)
```

```python
y_pred_gaussian = svc_gaussian.predict(X_test)

accuracy_gaussian = accuracy_score(y_test, y_pred_gaussian)
print("Accuracy:", accuracy_gaussian)
```

```python
#Sigmoid
svc_sigmoid = SVC(kernel='sigmoid')
svc_sigmoid.fit(X_train, y_train)
```

```python
y_pred_sigmoid = svc_sigmoid.predict(X_test)

accuracy_sigmoid = accuracy_score(y_test, y_pred_sigmoid)
print("Accuracy:", accuracy_sigmoid)
```
